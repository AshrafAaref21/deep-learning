### Creating Random Tensor

```python
random = tf.random.Generator.from_seed(42)
print(random)

random_1 = random.normal(shape=(3,2))
print('\n',50*'*','\n')
print(random_1)

random_2 = random.normal(shape=(3,2))
print('\n',50*'*','\n')
print(random_2)
```

### Shuffle The order of elements in a tensor

```python
not_shoffled = tf.constant([[10,7],
                            [3,4],
                            [2,5]])
```

```python
print(not_shoffled)
print('\n',50*'*','\n')
tf.random.shuffle(not_shoffled)

# Rerun and will shuffle it again
```

```python
print(not_shoffled)
print('\n',50*'*','\n')
tf.random.shuffle(not_shoffled,seed=42)

# Rerun and will shuffle it again
```

```python
print(not_shoffled)
print('\n',50*'*','\n')
tf.random.set_seed(42)
tf.random.shuffle(not_shoffled,seed=42)

# Won't shuffle it again with rerun
```

```python
print(not_shoffled)
print('\n',50*'*','\n')
tf.random.set_seed(42) # Global random seed
tf.random.shuffle(not_shoffled,seed=78) # operational level random seed

# Won't shuffle it again with rerun


# seed included in shuffle method is resetting the global seed but
# can't create the seed variable
```

### Matrix Multiplication
In Machine Learning, Matrix Multiplication is one of the most common tensor operations.
> matrix_1(m*n) == matrix_2(n*p)
>> #rows of mat1 == #cols of mat2

> output matrix has (m,p)

```python
tensor = tf.constant([[1,2],
                     [5,10]])
```

```python
tf.matmul(tensor,tensor)
```

```python
tensor * tensor # Normal item by item Multiplication
```

```python
tensor @ tensor # Matrix Multiplication (Dot Product)
```

```python
X = tf.constant([[1,2,3],
                 [4,5,6]])
Y = tf.constant([[2,5,10],
                 [1,3,5]])
```

```python
tf.tensordot(tf.transpose(X),Y,axes=1),tf.matmul(tf.transpose(X),Y)
```

```python
tf.tensordot(tf.reshape(X,shape=(3,2)),Y,axes=1)
```

### Tensor Aggregation

```python
E = tf.constant(np.random.randint(0,100,size=50))
tf.reshape(E,shape=(10,5))
```

```python
# Find Minimum
tf.reduce_min(E)
```

```python
# Find Maximum
tf.reduce_max(E)
```

```python
# Find Mean
tf.reduce_mean(E)
```

```python
# Find Sum
tf.reduce_sum(E)
```

```python
# Find Variance
import tensorflow_probability as tfp
tfp.stats.variance(E), tf.math.reduce_variance(tf.cast(E,dtype=tf.float32))
```

```python
# Find Standard Deviation
tf.math.reduce_std(tf.cast(E,dtype=tf.float32)) # Data Type has to be Float
```

#### Finding The Positional Maximum and Minimus

```python
F = tf.random.uniform(shape=[50])
F
```

```python
# Find The positional maximum
tf.argmax(F)
```

```python
F[tf.argmax(F)] == tf.reduce_max(F)
```

```python
# Find The minimum using the positional mimum index
F[tf.argmin(F)]
```

### One-hot Encoding Tensors

```python
ls = [0,1,2,3] # could be[red,green,blue,purple]
tf.one_hot(ls,depth=4) # 4 = len(ls)
```
